```python
import low

db = low.LanguagesOfTheWorld()


```

```python
for country in db.countries:

    speaker_counts = set()
    for language in country.speaker_counts:
        speaker_counts.add(language.language.part3)

    for language in country.languages:
        if language.part3 not in speaker_counts:
            print(country.label, language.label, language.part3)
```
